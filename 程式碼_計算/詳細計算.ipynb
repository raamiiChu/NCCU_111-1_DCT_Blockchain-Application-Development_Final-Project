```python
import pandas as pd
import numpy as np
import re
from typing import *
```

```python
menu = pd.read_excel("fx(hash) 作品清單.xlsx", index_col = 0)
before = pd.ExcelFile("先前作品.xlsx")
anniversary = pd.ExcelFile("一周年作品.xlsx")
after = pd.ExcelFile("之後作品.xlsx")
```

# 計算四種數據

```python
def cal(name: str, excel_file: pd.io.excel._base.ExcelFile) -> List[str and int]:
    result = []
    work = menu.loc[name]
    
    author = menu.loc[name]["author"]
    url = menu.loc[name]["url"]

    result.append(author)
    result.append(url)

    result.append(round((work["primVolumeTz"] / 10**6) * (work["splitsPrimary"] / 10**3) * 0.95, 4))
    
    result.append(round((work["secVolumeTz"] / 10**6) * (work["royalties"] / 10**3) * (work["splitsSecondary"] / 10**3), 4))

    data = excel_file.parse(name, index_col = "name", usecols = "C:H")
    gas_fee = data["gas_fee"]
    result.append(round((work["primVolumeTz"] / 10**6) / work["primVolumeNb"] + sum(gas_fee) / len(gas_fee), 4))
    
    result.append(round((work["secVolumeTz"] / 10**6) / work["secVolumeNb"], 4))

    return result
```

```python
sales_dict = {}

for index, name in enumerate(menu.index):
    sub_name = re.sub("[/\\[\\]]", " ", name)
    
    if index % 3 == 0:
        sales_dict[name] = cal(name, before)
    elif index % 3 == 1:
        sales_dict[name] = cal(name, anniversary)
    elif index % 3 == 2:
        sales_dict[name] = cal(name, after)
```

# 結果展示

```python
final_data = pd.DataFrame(sales_dict, index = ["作者", "網址", "一級市場收入", "二級市場收入", "一級市場平均售價", "二級市場平均售價"]).T
```
